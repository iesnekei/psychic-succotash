```python
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
```

```python
boston = load_boston()
```

```python
boston.data.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(boston['data'], boston['target'], test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
mean_x = X_train.mean(axis=0)
```

```python
std_x = X_train.std(axis=0)
```

```python
def norm_data(x):
    mean = x.mean(axis=0)
    std = x.std(axis=0)
    return (x - mean) / std
```

```python
X_test = norm_data(X_test)
```

```python
max_y = y_train.max()
min_y = y_train.min()
```

```python
y_train = y_train - min_y
y_train /= max_y
```

```python
y_test = (y_test - min_y) / max_y
```

```python
model = Sequential([
    Dense(512, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(512, activation='linear'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
                     loss='mse',
                     metrics=['mae'])

model.summary()
```

```python
print("On GPU:")
with tf.device("GPU:0"):
  
  history = model.fit(X_train, y_train,
  epochs=5,
  batch_size=20, validation_split=0.2)
```

```python
pred = model.predict(X_test).flatten()
```

```python
pred = pred * max_y + min_y
```

```python
y_test = y_test * max_y + min_y
```

```python
err = np.mean(abs(pred - y_test))
err
```

```python
for i in range(len(pred)):
    print("Сеть сказала: ", round(pred[i],2), ", а верный ответ: ", round(y_test[i],2), ", разница: ", round(pred[i] - y_test[i],2))
```

```python
results = {'level': [], 'neurons': [],  'mae': []}
```

```python
for i in range(15):
    level = np.random.choice(random_layers[0])
    n_n = np.random.choice(random_neurons[0])
    with tf.device("GPU:0"):
        inputs = Input(shape=(X_train.shape[1]), name='digits') 
        x=Flatten()(inputs)
        for i in range(level):
            #x = Dense(64, activation='relu', name='dense_'+str(i))(x) 
            x = Dense(n_n, activation='sigmoid', name='dense_'+str(i))(x) 
        outputs = Dense(1, activation='sigmoid' ,name='predictions')(x) 
        model = Model(inputs=inputs, outputs=outputs) 

        model.compile(optimizer='adam',
                            loss='mae',
                            metrics=['mse'])


  
        history = model.fit(X_train, y_train,
        epochs=5,
        batch_size=20, validation_split=0.2)

        pred = model.predict(X_test).flatten()
        pred = pred * max_y + min_y
        y_test = y_test * max_y + min_y
        err = np.mean(abs(pred - y_test))
        results['level'].append(level)
        results['mae'].append(err)
        results['neurons'].append(n_n)
        
        y_test = (y_test - min_y) / max_y

```
